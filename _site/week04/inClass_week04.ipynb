```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd

import ipywidgets
```

```python
michigan_file = '/Users/jnaiman/Downloads/michigan_lld.flt'
michigan = np.fromfile(michigan_file, dtype='f4').reshape((5365,4201))
# shape and data type come from the documentation of this file
# https://www.ngdc.noaa.gov/mgg/greatlakes/michigan.html
```

```python
michigan[michigan==-9999] = np.nan
```

```python
# 1. hardcode some options we want to make interactive
colormap_name = 'Blues'
take_log = True

if take_log:
    # compressing data into few colors
    norm = matplotlib.colors.SymLogNorm(10)
else:
    # linear normalization
    norm = matplotlib.colors.Normalize()

plt.imshow(michigan,cmap=colormap_name, norm=norm)
plt.colorbar()
plt.clim(-352,352)
plt.show()
```

```python
# 2. turn hard-coded options into a function
#colormap_name = 'Blues'
#take_log = True

def make_michigan_plot(colormap_name,take_log=True):
    # on a mac --> select lines --> COMMAND ]
    # windows: select lines --> CMD ]
    if take_log:
        # compressing data into few colors
        norm = matplotlib.colors.SymLogNorm(10)
    else:
        # linear normalization
        norm = matplotlib.colors.Normalize()

    plt.imshow(michigan,cmap=colormap_name, norm=norm)
    plt.colorbar()
    plt.clim(-352,352)
    plt.show()
```

```python
# 2.5 -- test your function!
make_michigan_plot('seismic')
```
